### **PROF_PRELIEVI**


All'interno trovi il codice che viene eseguito nell'extend_data


```python
import pandas as pd
```

```python
ext_period = pd.date_range(start='2023-10-01', end='2023-12-31', freq='D')
```

```python
add_back_start= '2022-10-01'
add_back_end =  '2022-12-31'
```

```python
df_profilo_prel_SQL = pd.read_excel('/content/prof_prelievi(22+23).xlsx')
```

```python
df_profilo_prel_SQL
```

```python
df_profilo_prel = df_profilo_prel_SQL[(df_profilo_prel_SQL['Dates']>=pd.to_datetime(add_back_start)) & (df_profilo_prel_SQL['Dates']<=pd.to_datetime(add_back_end))] #possibile errore
df_profilo_prel#['Profilo'] = 'C1A1'
```

```python
df_unique_values = df_profilo_prel[~df_profilo_prel.loc[:, df_profilo_prel.columns!='Dates'].duplicated()]

```

```python
df_ext_list = list()
for dt_ext in ext_period:
    temp_df = df_unique_values.copy()
    temp_df.Dates = dt_ext
    df_ext_list.append(temp_df)
df_ext = pd.concat(df_ext_list)
```

```python
df_profilo_prel_ext = pd.concat([df_profilo_prel_SQL, df_ext]).sort_values(by=['Profilo','Dates']) #possibile errore
df_profilo_prel_ext
```

```python
from decimal import Decimal
df_perc_prelievi_ext = pd.read_excel('/content/perc_prelievi.csv.xlsx')
```

```python
df_perc_prelievi_ext
```

```python
df_perc_prelievi_ext['c1%X1'] = Decimal(1)
df_perc_prelievi_ext['c1%X2'] = Decimal(1)
df_perc_prelievi_ext['c1%X3'] = Decimal(1)
df_perc_prelievi_ext.drop('timestamp_caricamento', inplace=True, axis=1)
df_perc_prelievi_ext.Data = (df_perc_prelievi_ext.Data + pd.offsets.DateOffset(years=1)).dt.strftime('%Y-%m-%d')
df_perc_prelievi_ext.set_index('Data', inplace=True)
```
